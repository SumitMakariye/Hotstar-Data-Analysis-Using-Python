###**1.Business Problem:**
The content strategy team wants to know whether to focus more on Movies or TV Shows on the platform.

###**Question:**

What is the count of Movies and TV Shows available on Hotstar?


```python
sns.countplot(x=df["type"], data=df, color="yellow")
plt.title("Count of Movies & TV Shows") #this gives us title to our chart
plt.show() #used to show the chart
```

Here we can saw that the count of Movies is more compared to TV Shows on the platform.
It means Hotstar has focused more on movie content rather than long-running TV shows or web series.

From a business point of view, Hotstar can plan to add more TV shows or exclusive web series to increase user engagement and keep viewers active for a longer time.

### **2.Business Problem:**

The marketing team wants to promote the most popular genres to attract new subscribers.

### **Question:**

Which genre has the highest number of titles on the platform?

```python
gb = df.groupby("type").agg({"hotstar_id":"count"})
gb
```

```python
plt.figure(figsize = (12,2)) #to incread the width and height of chart
sns.countplot(x = df["genre"], data=df, color="yellow")
plt.title("Count of genre")
plt.xticks(rotation=90)
plt.show()
```

Here we can saw that the Drama genre is the most popular on the platform, followed by Comedy and Romance.
It means people like to watch emotional and story-based content more compared to other genres.

From a business point of view, Hotstar can focus more on producing and promoting Drama and Comedy shows or movies, as they attract more audience and engagement.

###**3.Business Problem:**

The compliance and content curation team wants to ensure there is enough content for all age groups.

###**Question:**

How is the content distributed across different age ratings (U, UA, A)?

```python
gb = df.groupby("age_rating").agg({"hotstar_id":"count"})
gb
```

```python
plt.pie(gb["hotstar_id"], labels = gb.index, autopct = "%.2f%%")
plt.show()
```

Here we can saw that the UA 13+ category has the most content on the platform, followed by U and then UA 16+.
It means most of the content is suitable for teenagers and family audiences, with a balanced amount of kids-friendly and slightly mature content.

From a business point of view, this shows that Hotstar mainly targets a family and teen-based audience, and they can focus on creating more content in this category to keep engagement high.

### **4.Business Problem:**

The management team wants to understand how content production or acquisition has changed over time.

### **Question:**

How many Movies and TV Shows were released each year on Hotstar, and what is the trend over the years?

```python
gb = df.groupby("year").agg({"hotstar_id":"count"})
gb
```

```python
sns.lineplot(x = gb.index, y = gb["hotstar_id"])
plt.show()
```


### **5.Business Problem**
The content strategy team wants to identify which genres are dominated by Movies and which by TV Shows, to plan future content production and licensing decisions.

### **Business Question**

For each genre, how many Movies and TV Shows are available on the platform?


```python
plt.figure(figsize=(12,5))
sns.countplot(x="genre", hue="type", data=df, palette="Set2")
plt.title("Genre-wise Distribution of Movies and TV Shows")
plt.xticks(rotation=90)
plt.show()
```

Here we can see how many TV shows and movies are available for each genre on Hotstar. From the chart, it’s clear that the Drama genre has the highest number of both movies and TV shows. The Reality genre mainly consists of TV shows, while genres like Action, Comedy, and Romance have a larger number of movies.

This shows that Hotstar focuses heavily on Drama content overall, and also produces more Reality-based TV shows, whereas movie content is dominant in entertainment genres like Action and Comedy.

### **6.Business Problem:**

The content and compliance teams want to analyze how content is distributed across different age ratings and genres, to ensure that there is a good balance of content for all audience segments — from kids to adults.

### **Question**

Which genres are popular among different age groups on Hotstar?

```python
gb_genre_age = df.groupby(["age_rating", "genre"]).size().reset_index(name="count")
plt.figure(figsize=(12,5))
sns.barplot(x="age_rating", y="count", hue="genre", data=gb_genre_age)
plt.title("Genre Popularity by Age Rating")
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

Here we can see that most of the content on Hotstar belongs to the U/A 13+ age category.\
The Drama genre has the highest number of titles across almost all age groups.\
We can also see that Comedy and Action genres are more common in the U and U/A 13+ categories.\
There are fewer shows and movies in the A and U/A 7+ categories, which means Hotstar mainly focuses on teen and family-friendly content.
